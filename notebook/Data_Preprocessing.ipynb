```
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
true = pd.read_csv('/content/drive/My Drive/True.csv')
fake = pd.read_csv('/content/drive/My Drive/Fake.csv')
df = pd.read_csv('/content/drive/My Drive/Fake News/Project/scraped_articles.csv')
```

```
scraped = df
scraped["label"] = scraped["site"].apply(lambda x: 1 if(x == "breitbart" or x == "infowars" or x == "prntly" or x == "worldnewsdailyreport") else 0)
scraped.drop(labels=["link", "site"], axis = 1, inplace=True)
```

```
fake["label"] = 1
true["label"] = 0
```

```
print (fake.shape)
print (true.shape)
print (scraped.shape)
```

```
true["text"] = true['text'].apply(lambda x: x.partition("-")[2])
```

```
final_df =  true.append(fake)
```

```
final_df = final_df.drop(['date'], axis = 1)
```

```
scraped = scraped.drop(['published'], axis = 1)
```

```
final_df = final_df.append(scraped)
```

```
final_df = final_df.drop(['subject'], axis = 1)
```

```
final_df = final_df.dropna()
```

```
final_df.to_csv('final_data.csv')
```
